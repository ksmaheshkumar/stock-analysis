```python
# Required modules
import sqlite3
import pandas as pd
import numpy as np
conn = sqlite3.connect("stocks.db")
```

```python
df = pd.read_sql("SELECT date, symbol, adjclose FROM eod",conn, parse_dates='date', index_col='date')
df['adjclose'] = df['adjclose'].astype(float)
```

```python
np.log(df['adjclose']/df['adjclose'].shift())
```

```python
df['log_return'] = np.where(df['symbol']==df['symbol'].shift(), np.log(df['adjclose']/df['adjclose'].shift()), 'NaN')
```

```python
df.groupby('symbol').count()
```
